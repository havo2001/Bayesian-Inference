### **1. [2 pts]** 
Take priors $𝑓_0(𝜇) = 1$ and $𝑓_1(𝜇) \sim N(2, 𝜎^2 = 4)$ and find corresponding posterior densities $𝑓(𝜇|𝑋_𝑖)$ [1 pt each]. Plot both densities in one figure.

#### **Answer**
#### **a. $𝑓_0(𝜇) = 1$**
We know that $X = (X_{1},...,X_{n}) \quad where \quad X_{i} \sim f(X_{i}|μ)$:
<br>

$$  
f(X_{i}|μ) =  N(μ,𝜎^2) \quad where \quad 𝜎^2 = 1
$$
<br>

Likelihood:
$$  
\rightarrow f(X|μ) = \prod_{i=1}^{n}f(X_{i}|μ) = {(\frac{1}{\sqrt{2\pi𝜎^2}})}^n\exp(\frac{-1}{2𝜎^2}\sum_{i=1}^{n}{(X_{i}-μ)}^2)
$$
<br>

Prior:
$$
f_{0}(μ) = 1 
$$
<br>

The Posterior formula:
$$
f_0(μ|X) = \frac{f(X|μ).f_0(μ)}{f(X)}
$$
<br>

$$
→f_0(μ|X) \propto f(X|μ).f_0(μ) \propto exp(\frac{-1}{2𝜎^2}\sum_{i=1}^{n}{(X_{i}-μ)}^2)
$$
<br>

Denote the sample mean is $μ_1$:
$$
Sample \ sum: \quad \sum_{i=1}^{n}X_{i} = nμ_1  
$$
<br>

Then:
$$
f_0(μ|X) \propto exp(\frac{-1}{2𝜎^2}(nμ^2 - 2μ.nμ_1 + \sum_{i=1}^{n}X_{i}^2))
$$

<br>

Since we can forget about the constant terms respect to μ:
$$
f_0(μ|X) \propto exp(\frac{-n}{2𝜎^2}{(μ - μ_1)}^2)
$$
<br>

If we denote:
$$
𝜎'^2 = \frac{𝜎^2}{n}, \quad  μ' = μ_1
$$
<br>

Finally:
$$
f_0(μ|X) = N(μ', 𝜎') \quad where \quad 𝜎'^2 = \frac{𝜎^2}{n}, \quad  μ' = μ_1
$$
<br>
<br>
<br>
<br>






#### **b. $𝑓_1(𝜇) \sim N(2, 𝜎^2 = 4)$**
We know that $X = (X_{1},...,X_{n}) \quad where \quad X_{i} \sim f(X_{i}|μ)$:
<br>

$$  
f(X_{i}|μ) =  N(μ,𝜎^2) \quad where \quad 𝜎^2 = 1
$$
<br>

Likelihood:
$$  
f(X|μ) = \prod_{i=1}^{n}f(X_{i}|μ)
$$
<br>

Prior:
$$
f_1(𝜇) = \frac{1}{\sqrt{2\pi𝜎_0^2}}\exp(\frac{-1}{2𝜎_0^2}{(𝜇 - 𝜇_0)}^2) \quad where \ 𝜇_0 = 2,\ 𝜎_0^2=4
$$
<br>

The Posterior formula:
$$
f_1(μ|X) = \frac{f(X|μ).f_1(μ)}{f(X)}
$$
<br>

$$
→f_1(μ|X) \propto f(X|μ).f_1(μ) \propto \exp(\frac{-1}{2𝜎^2}\sum_{i=1}^{n}{(X_{i}-μ)}^2).\exp(\frac{-1}{2𝜎_0^2}{(𝜇 - 𝜇_0)}^2) \\ \propto exp(\frac{-1}{2𝜎^2}\sum_{i=1}^{n}{(X_{i}-μ)}^2 + \frac{-1}{2𝜎_0^2}{(𝜇 - 𝜇_0)}^2)
$$
<br>

Denote the sample mean is $μ_1$ and forget all the constants:
$$
Sample \ sum: \quad \sum_{i=1}^{n}X_{i} = nμ_1  
$$
<br>

Then:
$$
f_1(μ|X) \propto exp(\frac{-μ^2}{2}(\frac{n}{𝜎^2}+\frac{1}{𝜎_0^2}) + μ(\frac{nμ_1}{𝜎^2} + \frac{μ_0}{𝜎_0^2}))
$$
<br>

If we denote:
$$
𝜎'^2 = {(\frac{n}{𝜎^2}+\frac{1}{𝜎_0^2})}^{-1}, \quad  μ' = 𝜎'^2(\frac{nμ_1}{𝜎^2} + \frac{μ_0}{𝜎_0^2})
$$
<br>

Then:
$$
f_1(μ|X) \propto exp(\frac{-1}{2𝜎'^2}{(μ - μ')}^2)
$$
<br>
Finally:
$$
f_1(μ|X) = N(μ', 𝜎') \quad where \quad 𝜎'^2 = {(\frac{n}{𝜎^2}+\frac{1}{𝜎_0^2})}^{-1}, \quad  μ' = 𝜎'^2(\frac{nμ_1}{𝜎^2} + \frac{μ_0}{𝜎_0^2})
$$
<br>
<br>

#### **Visualize this possterior densities**

```python
# Sample properties:
mu = X.mean() # Sample mean
sigma = 1 # Sample standard deviation
var = 1 # Sample variance
n = 100
```

$$
f_0(μ|X) = N(μ', 𝜎') \quad where \quad 𝜎'^2 = \frac{𝜎^2}{n}, \quad  μ' = μ_1
$$
<br>
$$
=> 𝜎' = \frac{𝜎}{\sqrt{n}}
$$

```python
# Posterior of uniform prior:
mu_u = mu
sigma_u = sigma / math.sqrt(n)
posterior_u = sts.norm(loc = mu_u, scale = sigma_u)
print("Mean μ of the posterior:", mu_u)
print("Std σ of the posterior:", sigma_u)
posterior_u.rvs(10)
```

$$
f_1(μ|X) = N(μ', 𝜎') \quad where \quad 𝜎'^2 = {(\frac{n}{𝜎^2}+\frac{1}{𝜎_0^2})}^{-1}, \quad  μ' = 𝜎'^2(\frac{nμ_1}{𝜎^2} + \frac{μ_0}{𝜎_0^2})
$$
<br>
$$
=> 𝜎' = {(\frac{n}{𝜎^2}+\frac{1}{𝜎_0^2})}^{-1/2}
$$

```python
# Posterior of normal prior:
# Distribution of prior:
mu_0 = 2
var_0 = 4

# Distribution of posterior:
var_norm = (n/var + 1/var_0) ** -1
mu_norm = var_norm * (n * mu / var + mu_0 / var_0)
sigma_norm = math.sqrt(var_norm)
posterior_norm = sts.norm(loc = mu_norm, scale = sigma_norm)

print("Mean μ of the posterior:", mu_norm)
print("Std σ of the posterior:", sigma_norm)
posterior_norm.rvs(10)

```

```python
mu_ = np.linspace(4, 6, 100)
fig, ax = plt.subplots()

ax.plot(mu_, posterior_u.pdf(mu_), label='Uniform distribution prior')
ax.plot(mu_, posterior_norm.pdf(mu_), label='Normal distribution prior')
ax.set_title("Posterior densities")
ax.set_xlabel(r'Value of parameter $\mu$')
ax.set_ylabel("PDF")

ax.legend()
plt.show()

```

### **3. [3 pts]** 
 Let $𝜃 = \exp 𝜇$. Find the posterior density for $𝜃$ analytically [1 pt each] and by simulation [1 pt
total] using priors $𝑓_0
(𝜇)$ and $𝑓_1(𝜇)$. Plot calculated or estimated posterior densities together.

#### **Find the posterior density for 𝜃 analytically** 

#### **Answer**

Base on example of 11.3(s “All of Statistics”, p. 180), CDF of the posterior density of $θ$ is $F_0(θ|X)$:

<br>

$$ F_Θ(θ|X) = ℙ(Θ ≤ θ|X) = ℙ(\exp μ ≤ θ|X) = ℙ(μ ≤ \lnθ|X) = F_μ(\ln \theta|X) 
$$

<br>

Differentiate, we obtain the density of posterior density of $\theta$:
$$p(θ|X) = f(\ln θ|X).\frac{1}{θ} \quad \text{where } f(\ln θ|X) \text{ is the posterior density of } μ $$
<br>









```python
def posterior_analytical(posterior, theta_):
  return posterior.pdf(np.log(theta_)) / theta_
```

#### **Find the posterior density for 𝜃  by simulation** 

#### **Answer**

Base on example of 11.4(s “All of Statistics”, p. 181), we can simulate the posterior density of $θ$ by:


1.   First use the distribution of posterior of $\mu$ to generate k points (we obtained above).
2.   Take the exponent on these k points, then the new k points are the simulation of the posterior density of $\theta$.










```python
k_ = 1000
```

#### **$f_0(\mu) = 1$**

```python
mu_X_uniform = posterior_u.rvs(k_)
theta_X_uniform = np.exp(mu_X_uniform)

```

#### **$𝑓_1(𝜇) \sim N(2, 𝜎^2 = 4)$**

```python
mu_X_norm = posterior_norm.rvs(k_)
theta_X_norm = np.exp(mu_X_norm)

```

#### **Plot two ways value together**

```python
theta_ = np.linspace(min(theta_X_uniform.min(), theta_X_norm.min()) - 50, max(theta_X_uniform.max(), theta_X_norm.max()) + 50, 1000)
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 6))

ax1.plot(theta_, posterior_analytical(posterior_u, theta_), label="Analytical way", color="r")
sns.histplot(theta_X_uniform, stat= "density", label="Estimation way", ax=ax1)
ax1.set_title("Prior f0 = 1")

ax2.plot(theta_,  posterior_analytical(posterior_norm, theta_), label="Analytical way",color="r")
sns.histplot(theta_X_norm, stat= "density", label="Estimation way", ax=ax2)
ax2.set_title("Prior f1 ~ N(2, 4)")

ax1.legend()
ax2.legend()

plt.show()
```

### **4. [2 pts]** 
Find a 95% posterior interval (PI) for both $𝜇$ and $𝜃$ using the results of previous simulations.

```python
def posterior_interval_simulation(simulation_values):
  return (np.quantile(simulation_values, 0.025), np.quantile(simulation_values, 0.975))
```

```python
# Posterior interval(PI) for μ:
print("95%% posterior interval (PI) for μ with prior f_0:  %.3f, %.3f" % posterior_interval_simulation(mu_X_uniform))
print("95%% posterior interval (PI) for μ with prior f_1:  %.3f, %.3f" % posterior_interval_simulation(mu_X_norm))
# Posterior interval(PI) for θ:
print("95%% posterior interval (PI) for θ with prior f_0:  %.3f, %.3f" % posterior_interval_simulation(theta_X_uniform))
print("95%% posterior interval (PI) for θ with prior f_1:  %.3f, %.3f" % posterior_interval_simulation(theta_X_norm))
```

### **5. [3 pts]** 
 [2 pts total] Obtain MAP estimates for $𝜇$ and $𝜃$. Compare them to MLEs $\hat{𝜇}_{𝑀𝐿𝐸}$ and $\hat{\theta}_{MLE} $ of

#### **Answer**
#### **Compute MLE:**
<br>

$$\hat{𝜇}_{MLE}=argmaxf(X|\mu)$$
<br>

##### **Consider posterior of $\mu$:**

$$f(X|\mu)=\prod_{i=1}^{n}f(X_{i}|μ) = {(\frac{1}{\sqrt{2\pi𝜎^2}})}^n.\exp(\frac{-1}{2𝜎^2}\sum_{i=1}^{n}{(X_{i}-μ)}^2)={(\frac{1}{\sqrt{2\pi𝜎^2}})}^n.exp(\frac{-1}{2𝜎^2}(nμ^2 - 2\mu.\sum_{i=1}^{n}X_{i} + \sum_{i=1}^{n}X_{i}^2))$$

<br>

Differentiate respect to $\mu$, $f(X|\mu)$ get maximized when:
<br>

$$\mu = \frac{\sum_{i=1}^{n}X_{i}}{n}=\mu_1 \quad \text{(Sample X mean)}$$
$$=> \hat{\mu}_{MLE} = \frac{\sum_{i=1}^{n}X_{i}}{n}=\mu_1$$
<br>
<br>


##### **Consider posterior of $\theta$ when $\theta=\exp(\mu)$:**
Then:
$$\ln \theta = \mu $$
<br>

$$f(X|\ln \theta)=\prod_{i=1}^{n}f(X_{i}|\mu) $$

Using the same result from above:
<br>

$$=> \hat{\theta}_{MLE} = \exp({\frac{\sum_{i=1}^{n}X_{i}}{n}})=\exp \mu_1$$












#### **Obtain MAP estimates**




#### **Answer**
#### **Compute MAP:**
<br>

$$\hat{𝜇}_{MAP}=argmaxf(\mu|X) \quad \text{(maximum of posterior density)}$$ 
<br>
$$\hat{\theta}_{MAP}=argmaxp(\theta|X) \quad \text{(maximum of posterior density)}$$













```python
# Obtain MAP via the estimation of posterior of mu and theta
# First getting pdf of posterior
posterior_pdf_mu_uniform = posterior_u.pdf(mu_X_uniform)
posterior_pdf_mu_norm = posterior_norm.pdf(mu_X_norm)
posterior_pdf_theta_uniform = posterior_analytical(posterior_u, theta_X_uniform) 
posterior_pdf_theta_norm = posterior_analytical(posterior_norm, theta_X_norm) 

# Getting the parameter when the pdf of posterior gets maximum, it is MAP value
mu_map_uniform = mu_X_uniform[posterior_pdf_mu_uniform.argmax()]
mu_map_norm = mu_X_norm[posterior_pdf_mu_norm.argmax()]
theta_map_uniform = theta_X_uniform[posterior_pdf_theta_uniform.argmax()]
theta_map_norm = theta_X_norm[posterior_pdf_theta_norm.argmax()]

print("MAP estimate for μ with f0:", mu_map_uniform)
print("MAP estimate for μ with f1:", mu_map_norm)
print("MAP estimate for θ with f0:", theta_map_uniform)
print("MAP estimate for θ with f1:", theta_map_norm)
```

### **6. [2 pts]** 
 Compute frequentist 95% confidence interval for $𝜇$ [1 pt total] and compare it with PIs from 
question 5 [1 pt total]. Interpret the results.

```python
def ci_computation(x :np.ndarray, n):
  alpha = 0.05
  z_value = sts.norm().ppf(alpha/2)
  lbd_hat = x.mean()
  se_hat = np.sqrt(lbd_hat / n)
  ci_lower = lbd_hat + se_hat * z_value
  ci_upper = lbd_hat - se_hat * z_value
  return ci_lower, ci_upper
```

```python
n
```

```python
ci_computation(X,n)
```

```python
print("95%% confident interval (CI) for μ:  %.3f, %.3f" % ci_computation(X,n))
# Posterior interval(PI) for μ:
print("95%% posterior interval (PI) for μ with prior f_0:  %.3f, %.3f" % posterior_interval_simulation(mu_X_uniform))
print("95%% posterior interval (PI) for μ with prior f_1:  %.3f, %.3f" % posterior_interval_simulation(mu_X_norm))
```

#### **Answer**
The 95% confidence interval (CI) is wider then the 95% posterior interval for $\mu$ in both priors.
#### **Interpret the results.**
*The frequentist approach uses a "confidence interval" - a range of values designed to include the true value of the parameter with some minimum probability, say 95% whereas "posterior interval" comes from Bayesian approaches.*
* As we can see, the CI interval is wider then the PI interval, it means that the PI interval is more reliable. Since our sample is not large enough and using Bayesian method with a prior knowledge gives more accuracy.
* Since the CI includes the true value of the parameter, and the parameter is fixed so the confidence interval is random (it depends on the random sample).
* The BI are based on the idea that the estimated parameters are random variables with a distribution, therefore BI is simply an interval.
